```python
GDP_correct=GDP.copy()
for index, row in GDP_correct.iterrows():
  if row['Aasta']>2016:
    GDP_correct = GDP_correct.drop(index)

GDP_correct
```

```python
crime_correct=crimes.copy()
for index, row in crime_correct.iterrows():
  if row['Aasta']>2016:
    crime_correct = crime_correct.drop(index)

crime_correct
```

```python
df=crime_correct.copy()
df.drop(df.columns[3:20], axis=1, inplace=True)
#df.drop(columns=[df.columns[3:19]])
df
```

```python
for index, row in df.iterrows():
  if row['Kuriteo aste/liik'].startswith('..'):
    df = df.drop(index)
  elif row['Kuriteo aste/liik'].endswith('astme kuriteod') or row['Kuriteo aste/liik'].endswith('kokku') :
    df = df.drop(index)
df
```

```python
df['Kokku'] = df.groupby(['Kuriteo aste/liik'])['Kogu Eesti'].transform('sum')
pie_chart_df = df.drop_duplicates(subset=['Kuriteo aste/liik']).drop(columns=['Kogu Eesti', 'Aasta'])
pie_chart_df
```

```python
muu=0
for index, row in pie_chart_df.iterrows():
  if row['Kokku']<=3400:
    muu+=row['Kokku']
    pie_chart_df = pie_chart_df.drop(index)
muu_rida = {'Kuriteo aste/liik': 'Muu', 'Kokku': muu}
pie_chart_df=pie_chart_df.append(muu_rida, ignore_index=True)
pie_chart_df
```

```python
import matplotlib.pyplot as plt
#pie_chart_df=pie_chart_df['Kokku'].astype(int)
labels=pie_chart_df['Kuriteo aste/liik'].tolist()
explode = (0.05,0.05,0.05,0.05,0.05,0.05,0.05,0.05,0.05,0.05,0.05)
plot=pie_chart_df.plot.pie(y='Kokku', figsize=(10, 10), labels=labels, autopct='%1.0f%%', colors=['#F2BB66', '#FFCBA4', '#FFA500', '#FF8674', '#FBE7A1', '#FAF0DD'], legend=False,explode = explode)
plot.set_ylabel('')
centre_circle = plt.Circle((0,0),0.35,fc='white')
fig = plt.gcf()
fig.gca().add_artist(centre_circle)
# Equal aspect ratio ensures that pie is drawn as a circle
plt.tight_layout()
plt.show()
```
